```python
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException 
import time
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.DataFrame()
```

```python
url = 'https://www.alljobs.co.il/SearchResultsGuest.aspx?page='
page = 1
driver =webdriver.Chrome()
driver.get(url + str(page))
driver.implicitly_wait(5)

```

```python
job_elements = driver.find_elements(By.CLASS_NAME,'job-content-top-title')
```

```python
for page in range(2,4):
    for element in job_elements:
        job_title = element.find_element(By.TAG_NAME,'h3').text
        link = element.find_element(By.TAG_NAME,'a').get_attribute('href')
        row = pd.Series({"title":job_title,'link':link})
        df = pd.concat([df, row.to_frame().T], ignore_index=True)

    url = 'https://www.alljobs.co.il/SearchResultsGuest.aspx?page='
    driver.get(url + str(page))
    driver.implicitly_wait(5)
    job_elements = driver.find_elements(By.CLASS_NAME,'job-content-top-title')

df
```

```python
driver =webdriver.Chrome()
for i in range(len(df)):
    driver.get(df.at[i,'link'])
    driver.implicitly_wait(1)
    element = driver.find_element(By.CLASS_NAME,"job-content-top")
    try:    
        company = element.find_element(By.CLASS_NAME,'T14').find_element(By.TAG_NAME,'a').text
    except:
        company = 'חברה חסויה'
    df.at[i,'company']=company
    print(i, ' in ', len(df))
    
```
